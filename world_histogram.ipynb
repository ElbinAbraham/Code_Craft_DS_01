```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv("world_population.csv")
%matplotlib inline
```

```python
sns.set_style("white")

plt.show()
```

```python

_, axs = plt.subplots(2,1,figsize=(20,16))
plt.tight_layout(pad=7.0)

sns.barplot(x=data["Country/Territory"],y=data["2020 Population"],hue=data["Country/Territory"],order=data.sort_values("2020 Population",ascending=True)["Country/Territory"][:11],ax=axs[0],palette='rainbow', saturation=1,edgecolor = "#1c1c1c", linewidth = 4)
axs[0].set_yscale("linear")
axs[0].set_title("Least Populated Countries In 2020",fontsize=25)
axs[0].set_xlabel("\nCountry",fontsize=20)
axs[0].set_ylabel("Population",fontsize=20)
axs[0].set_xticklabels(axs[0].get_xticklabels(),rotation = 12)


sns.barplot(x=data["Country/Territory"],y=data["2020 Population"],hue=data["Country/Territory"],order=data.sort_values("2020 Population",ascending=False)["Country/Territory"][:11],ax=axs[1],palette='rainbow', saturation=1,edgecolor = "#1c1c1c", linewidth = 4)
axs[1].set_yscale("log")
axs[1].set_title("Top Populated Countries In 2020",fontsize=25)
axs[1].set_xlabel("\nCountry",fontsize=20)
axs[1].set_ylabel("Population",fontsize=20)
axs[1].set_xticklabels(axs[1].get_xticklabels(),rotation = 0)

sns.despine(left=True, bottom=True)
plt.show()
```
